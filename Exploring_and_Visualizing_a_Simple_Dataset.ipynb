
# Task 1: Exploring and Visualizing a Simple Dataset

## Objective
The goal of this project is to explore and visualize the Iris dataset using Python libraries such as Pandas, Matplotlib, and Seaborn. The analysis helps understand data distributions, feature relationships, and potential outliers.



```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset("iris")
df.head()
```

```python
print("Shape of dataset:", df.shape)
print("Column names:", list(df.columns))
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
plt.figure(figsize=(7,5))
sns.scatterplot(data=df, x="sepal_length", y="petal_length", hue="species")
plt.title("Sepal Length vs Petal Length by Species")
plt.show()
```

```python
df.hist(figsize=(10,8), bins=20, edgecolor="black")
plt.suptitle("Feature Distributions (Histograms)")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,6))
df_melt = df.melt(id_vars="species", var_name="feature", value_name="value")
sns.boxplot(data=df_melt, x="feature", y="value", hue="species")
plt.title("Box Plots of Features by Species (Outlier Detection)")
plt.xticks(rotation=20)
plt.tight_layout()
plt.show()
```

## Conclusion

### Key Insights

- The Iris dataset contains 150 observations and 5 columns.
- Petal length and petal width show strong separation among species.
- Setosa is clearly distinguishable from Versicolor and Virginica.
- Histograms indicate the distribution of each feature.
- Box plots reveal a few potential outliers, particularly in sepal measurements.
- The dataset is clean and suitable for machine learning applications.
